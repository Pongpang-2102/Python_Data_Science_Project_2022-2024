## 1.1 จงเขียนตาราง ANOVA

```python
# 1.1.1 import library

## 1.1.1.1 import pandas library to create and manipulate dataframe
import pandas as pd
import numpy as np

## 1.1.1.2 import stats model to find an ANOVA
import statsmodels.api as sm
import statsmodels.formula.api as smf

# 1.1.1.3 import sklearn to predict from our model
from sklearn.linear_model import LinearRegression
```

```python
# 1.1.2 create an independent variable (X)
# and dependemt variable (Y)
X_input =  [1,4,11,11,2,2,1,5,10,8]
Y_input =   [41,59,90,88,52,53,53,63,87,74]

# to view our input
print("X input : " ,X_input )
print("Y input : " ,Y_input )

```

```python
# 1.1.3 create store as dictionary
raw_data  = {
    "i" :  [i for i in range(1,11,1) ]    ,
    "X" : X_input ,
    "Y" : Y_input
}
```

```python
# 1.1.4 create dataframe from our created dictioanry
df_topic01 = pd.DataFrame(raw_data)
df_topic01 
```

```python
# 1.1.4 create X dataframe 
X = df_topic01['X']
X
```

```python
# 1.1.5 create Y dataframe 
Y= df_topic01['Y']
Y
```

```python
# 1.1.6 put X , Y into OLS Regression and show the ANOVA
X_add_cons = sm.add_constant(X)
est = sm.OLS(Y,X_add_cons ).fit()
print( est.summary() )  # our ANSWER
```

## 1.2	จงหาช่วงความเชื่อมั่น 99% ของ E(Yh) และ Yh เมื่อกำหนดค่า  Xh = 7

### 1.2.1 จงหาช่วงความเชื่อมั่น 99% ของ E(Yh)

```python
# 1.2.1 to view beta_0 and beta_1 at 99%Ci for this info
est_99 = est.conf_int(alpha=0.025, cols=None)
est_99

# 0 is Beta_0 and 1 is Beta_1 Respectively
```

### 1.2.2 จงหา Yh เมื่อกำหนดค่า Xh = 7

```python
# calculate prediced Yh of Xh = 7 from same condition with 1.2.1

#X_input =  [1,4,11,11,2,2,1,5,10,8]

## 1.2.2.1 start with create model
X = np.array(X_input).reshape((-1,1))
y = df_topic01['Y']
```

```python
# 1.2.2.2 create our Linear Regression  Model & fit our X,y to model
model = LinearRegression()
model.fit(X,y)
```

```python
# 1.2.2.3 define xh = 7
Xh = 7

# 1.2.2.4 try to predict model and see our answer
y_predicted = model.predict([[Xh]])

print("if our Xh = ",Xh,"\ny will be = ",y_predicted[0])
```

## 2.2 Creating Multiple Linear Regression Model with Explanation

```python
## 1.2.2.2 speficy target_y column : Heating_Oil
target_y = oil_df.Heating_Oil


## ## 1.2.2.3 speficy feature col (or X variable) 

feature_x = oil_df[[ 'Insulation' , 'Temperature' , 'Num_Occupants'  , 'Avg_Age', 'Home_Size']]

# we will be using our feature column 
# (Insulation , Temperature ,Num_Occupants , Avg_Age , Home_Size )
# to predict our target column(Heating_oil teamperature ) or target_Y

```

```python
# 1.2.2.4 create our Linear Regression  Model & fit our X,y to model
model = LinearRegression()
model.fit(feature_x,target_y)
```

```python
## 1.2.2.5 to view model info
r_sq = model.score(feature_x, target_y)
y_intercept = model.intercept_
slope = model.coef_

print("R square = ",r_sq)
print("intercept = ",y_intercept)
print("slope\nSlop for Insulation : ",slope[0])
print("Slop for Temperature  : ",slope[1])
print("Slop for Num_Occupants : ",slope[2])
print("Slop for Avg_age : ",slope[3])
print("Slop for Home_size ",slope[4])

```

## 2.3 Model Mean Square Error

```python
# 2.3.1 Start with finding y prediction
predicted_y = model.predict(feature_x)
predicted_y
```

```python
# 2.3.2 import neccessary library
from sklearn.metrics import mean_squared_error
```

```python
# 2.3.3 use target with predicted y to find Mean squared error
print("mean Squared Error = " ,mean_squared_error(target_y,predicted_y) )
```
